# GLOBAL ENERGY CONSUMPTION ANALYSIS

## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy6; as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv("global_energy_consumption_dataset.csv")

df.head()
```

## BASIC INFORMATION

```python
print("Rows, columns:",df.shape)

print(df.info())

df.describe()
```

## MISSING VALUES

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df_cleaned = df.dropna()

print("After cleaning, missing values:")
print(df_cleaned.isnull().sum())
```

## Univariative Analysis (Single Column)

```python
# Histogram
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned["Total_Energy_Consumption_TWh"], bins=50, kde=True)
plt.title("Distribution of Total Energy Consumption (TWh)")
plt.xlabel("Energy Consumption (TWh)")
plt.ylabel("Frequency")
plt.show()

# Boxplot
plt.figure(figsize=(8,3))
sns.boxplot(x=df_cleaned["Total_Energy_Consumption_TWh"])
plt.title("Boxplot of Total Energy Consumption")
plt.show()
```

## Biviariate Analysis (Two Columns)

```python
# Scatter plot: GDP vs Energy Consumption
plt.figure(figsize=(8,5))
sns.scatterplot(x="GDP_BillionUSD", y="Total_Energy_Consumption_TWh", data=df_cleaned)
plt.title("GDP vs Total Energy Consumption")
plt.xlabel("GDP (Billion USD)")
plt.ylabel("Total Energy Consumption (TWh)")
plt.show()

# Correlation heatmap (numeric columns only)
numeric_df = df_cleaned.select_dtypes(include=[np.number])   # only numeric data
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

## Outlier Detections

```python
# IQR method for Total Energy Consumption
Q1 = df_cleaned["Total_Energy_Consumption_TWh"].quantile(0.25)
Q3 = df_cleaned["Total_Energy_Consumption_TWh"].quantile(0.75)
IQR = Q3 - Q1

outliers = df_cleaned[(df_cleaned["Total_Energy_Consumption_TWh"] < (Q1 - 1.5 * IQR)) | 
                      (df_cleaned["Total_Energy_Consumption_TWh"] > (Q3 + 1.5 * IQR))]

print("Number of outliers in Energy Consumption:", outliers.shape[0])
```
